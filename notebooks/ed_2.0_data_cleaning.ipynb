```python
# Creates a dictionary from df_train where the keys are columns names and the values are lists of the index positions that have the same column name
df_train_duplicate_cols = {}
for idx, col in enumerate(df_train.columns):
    if col in df_train_duplicate_cols:
        df_train_duplicate_cols[col].append(idx)
    else:
        df_train_duplicate_cols[col] = [idx]
```

```python
# Uses the dictionary made from df_train to check if columns that share the same name are duplicates or not 
for cols, indices in df_train_duplicate_cols.items():
    if len(indices) > 1:
        print(f"Column {cols} appears at index positions: {indices}")
        true_false = df_train.iloc[:, indices[0]] == df_train.iloc[:, indices[1]]
        true_false_all = true_false.all()
        if true_false_all == np.False_:
            print(f"The columns at index {indices[0]} and {indices[1]} are not duplicates.")
            print(f"At least one row of data in the column at index {indices[0]} is different than the column at index {indices[1]}.")
        else:
            print(f"The column at index {indices[0]} is different than the column in {indices[1]}")
            print(f"All data in the column at index {indices[0]} is the same as the data the column in {indices[1]}")
        print("---")
```

```python
# Creates a dictionary from df_test where the keys are columns names and the values are lists of the index positions that have the same column name
df_test_duplicate_cols = {}
for idx_test, col_test in enumerate(df_test.columns):
    if col_test in df_test_duplicate_cols:
        df_test_duplicate_cols[col_test].append(idx_test)
    else:
        df_test_duplicate_cols[col_test] = [idx_test]
```

```python
# Uses the dictionary made from df_test to check if columns that share the same name are duplicates or not
for cols_test, indices_test in df_test_duplicate_cols.items():
    if len(indices_test) > 1:
        print(f"Column {cols_test} appears at index positions: {indices_test}")
        test_true_false = df_test.iloc[:, indices_test[0]] == df_test.iloc[:, indices_test[1]]
        test_true_false_all = test_true_false.all()
        if test_true_false_all == np.False_:
            print(f"The columns at index {indices_test[0]} and {indices_test[1]} are not duplicates.")
            print(f"At least one row of data in the column at index {indices_test[0]} is different than the column at index {indices_test[1]}.")
        else:
            print(f"The column at index {indices_test[0]} is different than the column in {indices_test[1]}")
            print(f"All data in the column at index {indices_test[0]} is the same as the data the column in {indices_test[1]}")
        print("---")
```

- I plan to encode most, if not all, the categorical data

- For an XGBoost model, I would encode the data now

- For a model that benefits from scaling, such as a neural network, I would wait until after the missing data is handled to do this
- If I wait until after the missing data is handled, I can scale the numerical columns and encode the categorical columns at the same time using a column transformer
- This will be less arduous, because the categorical columns will still have the object datatype.
- If I encoded the categorical columns before scaling the numerical columns, all the columns would have the float64 or int64 datatype and I would have to define the columns for scaling by hand
- By doing both transformation at the same type I can define the columns for encoding and the columns for scaling using the select_dtype method

```python
# Function to identify and drop highly correlated features, as well as features with low variance
def preprocess_corr_var(df, correlation_threshold=0.9, variance_threshold=0.01):
    # Calculate the correlation matrix
    corr_matrix = df.corr().abs()
    
    # Identify pairs of highly correlated features
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    high_corr_pairs = upper_triangle.stack().reset_index()
    high_corr_pairs.columns = ['Feature_1', 'Feature_2', 'Correlation']
    high_corr_pairs = high_corr_pairs[high_corr_pairs['Correlation'] > correlation_threshold]
    
    # Create a set to keep track of features to drop
    features_to_drop = set()
    
    # Iterate over each pair of highly correlated features
    for _, row in high_corr_pairs.iterrows():
        feature_1 = row['Feature_1']
        feature_2 = row['Feature_2']
        # Drop the feature with the lower variance
        if df[feature_1].var() < df[feature_2].var():
            features_to_drop.add(feature_1)
        else:
            features_to_drop.add(feature_2)

    # Drop the identified features
    num_corr_features_to_drop = len(features_to_drop)
    print(f"{num_corr_features_to_drop} highly correlated features were dropped from the dataset")
    features_to_drop_indices = [df.columns.get_loc(feature) for feature in features_to_drop]
    df_corr_dropped = df.drop(columns=list(features_to_drop))

    # Initialize the VarianceThreshold selector
    selector = VarianceThreshold(threshold=variance_threshold)

    # Fit the selector to the data and transform the data
    df_corr_var_dropped = pd.DataFrame(selector.fit_transform(df_corr_dropped), 
                                       columns=df_corr_dropped.columns[selector.get_support()]) # returns a boolean array indicating which features are retained (True) and which are removed (False) after applying the variance threshold. This boolean array can be used to filter the columns of the original DataFrame to keep only the features that meet the variance threshold.
    
    low_variance_dropped_indices = [i for i, x in enumerate(selector.get_support()) if not x]
    # Calculate the number of features dropped due to low variance
    num_features_dropped_low_variance = len(low_variance_dropped_indices)
    print(f"{num_features_dropped_low_variance} low variance features were dropped from the dataset")
    
    return df_corr_var_dropped, corr_features_to_drop_indices, low_variance_dropped_indices
```

### Data Source and Reliability
What is the origin of the data? How was it obtained? For quality assurance purposes, it's relevant to know how reliable the data is. This will help you deal with any potential data-entry errors or determine how trustworthy your eventual model will be.

- The data is gathered by Vesta's (the sponsor of the Kaggle competition) fraud protection system and digital security partners
- A transaction is labeled as fraud (isFraud=0) if there is a reported chargeback on a given card. Transactions occuring after the fraudulent one are also labeled as fraud if they are directly linked to the fraudulent transaction by attributes such as user account, email address or billing address.
- If none of above is reported and found beyond 120 days, then a transaction is labeled as not fraud (isFraud=0).
- However, it is noted, that in the real world, fraudulent activity might not be reported, e.g. cardholder was unaware of fraudulent activity on their card, or did not report it in time (beyoned the claim period, etc.) In such cases, supposed fraud might be labeled as legit, but it would be practically impossible to know. From a representative of Vesta - "We think they're unusual cases and a negligible portion."
- In regard to the dataset, this means that it is difficult to trust the complete accuracy of the labeling and does raise questions about how reliable an eventual model will be, though the sponsor doesn't think mislabeled data make up a significant slice of the data
